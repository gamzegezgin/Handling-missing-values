```python
import pandas as pd
import numpy as np

#verileri oku
nlf_data = pd.read_csv('NFL Play by Play 2009-2017 (v4).csv', low_memory=False)
sf_permits = pd.read_csv('Building_Permits.csv',  low_memory=False)
```

```python
np.random.seed(0)

#bunun sebebini henüz kavrayamadım
```

```python
nlf_data.sample(5)
```

```python
sf_permits.sample(5)
```

```python
#her kolondaki eksik değer tespiti için tanımlama yapalım
missing_values_count = nlf_data.isnull().sum()

#sutunlardaki eksik veri sayılarını öğrenelim

missing_values_count[:]
```

```python
#toplam kaç eksik verimiz oldugunu kontrol edelim

total_cells = np.prod(nlf_data.shape)
total_missing = missing_values_count.sum()

#eksik verilerin oranına bakalım

missing_percentage = (total_missing / total_cells) * 100
print(missing_percentage)


#nlf_data.shape komutu, nlf_data adlı DataFrame'in satır ve sütun sayısını bir tuple olarak döndürür. Örneğin (20,100)
#eğer nlf_data.shape[1] şeklinde yazarsak;
#shape[0]: Satır sayısını verir (ilk öğe).
#shape[1]: Sütun sayısını verir (ikinci öğe).
#np.prod(nlf_data.shape) komutu, bu iki değerin çarpımını alır. Yani, toplam hücre sayısını (100 * 20 = 2000 hücre) hesaplar.

#missing_values_count.sum() ile her sutunda bulunan boş hücrelerin sayısı toplanır
```

```python
#aynı şeyleri sf_permits için yapacağız

missing_values_count1 = sf_permits.isnull().sum()

missing_values_count1[0:10]
```

```python
total_cells1 = np.prod(sf_permits.shape)
total_missing1 = missing_values_count1.sum()

missing_percentage1 = (total_missing1 / total_cells1) * 100
print(missing_percentage1)
```

```python
nlf_data.dropna()
#1 tane bile eksik değer olan satırı sildiği için tüm satırlar silindi
```

```python
columns_with_na_dropped = nlf_data.dropna(axis = 1)
columns_with_na_dropped.head()

# axis=0 (default): Satırları ifade eder. Eğer axis=0 olarak belirtilirse, eksik değerlere sahip olan satırlar kaldırılır.
# axis=1: Sütunları ifade eder. Eğer axis=1 olarak belirtilirse, eksik değerlere sahip olan sütunlar kaldırılır.
```

```python
# eksik verilerin oldugu sutunları sildikten sonra ne kadar veri kaybettiğimizi kontrol edelim

print("Orijinal veri seti sütun sayısı : %d \n" % nlf_data.shape[1])

print("Eksik verileri çıkarıldıktan sonraki sütun sayısı : %d" % columns_with_na_dropped.shape[1])
```

```python
#şimdi aynı seyler sf_permits veri seti için yapacağız 

columns_with_na_dropped1 = sf_permits.dropna(axis=1)
columns_with_na_dropped1.head()
```

```python
print("Orginal veri setinin sutunları : %d \n" % sf_permits.shape[1])
print("Eksik değerler silindikten sonra sf_permits veri setindeki  sütün sayısı : %d" % columns_with_na_dropped.shape[1])
```

```python
#NLF veri setinden bir alt küme veri seti alıyoruz
#tüm satırların dahil edildiği, 'EPA' sütunundan, 'Season' sütununa kadar 

subset_nlf_data = nlf_data.loc[:, 'EPA':'Season'].head()
subset_nlf_data
```

```python
#subset_nlf_data.fillna(0): Tüm eksik değerleri 0 ile doldurur

subset_nlf_data.fillna(0)
```

```python
#subset_nlf_data.fillna(method='bfill', axis=0).fillna("0"):
#İlk olarak, NaN değerleri geri doldurma yöntemiyle (sonraki geçerli değerle) doldurur.
#Geri doldurma sonrasında hala NaN olan değerleri "0" (string) ile doldurur.
#axis=0: Bu, işlemin satır bazında yapılacağını belirtir. Yani sütunlar yukarıdan aşağıya doğru doldurulur.Satır satır ilerlemek için yukarıdan aşağıya olmalı bö

#subset_nlf_data.fillna(method = 'bfill', axis = 0).fillna("0") (eski kullanım)
subset_nlf_data.bfill(axis=0).fillna("0")
```

```python
#aynı şeyleri sf_permits için de yapalım ama bu kez alt küme oluşturmadan tüm veri setini dolduracağız

sf_permits.bfill(axis=0).fillna("0").head()
```
